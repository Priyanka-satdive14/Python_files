### Type of Machine Learning
Machine Learning (ML) is a subset of Artificial Intelligence (AI) that involves training algorithms to learn from data and make predictions or decisions without being explicitly programmed. ML can be broadly categorized into three main types, based on the nature of the learning process and the availability of labeled data. An additional category, Reinforcement Learning, is often discussed separately due to its unique approach.

#### Supervised Learning
In supervised learning, the model is trained using labeled data, meaning the input data is paired with the correct output (target labels). The goal is to learn a mapping function from inputs to outputs.

**Key Characteristics:**
- Labeled dataset (input-output pairs are known).
- The model's performance is evaluated by how well it predicts the outputs for unseen inputs.

**Applications:**
- **Classification:** Predicting discrete labels (e.g., spam email detection, image classification).
- **Regression:** Predicting continuous values (e.g., house price prediction, stock price forecasting).

**Examples of Algorithms:**
- Linear Regression
- Logistic Regression
- Decision Trees
- Support Vector Machines (SVM)
- Neural Networks (for tasks like image recognition).
#### Unsupervised Learning
In unsupervised learning, the data does not have labeled outputs. The algorithm tries to learn the underlying structure or distribution in the data.

**Key Characteristics:**
- No labels provided.
- Focus is on discovering patterns or grouping data points based on similarities.

**Applications:**
- **Clustering:** Grouping similar data points (e.g., customer segmentation, market basket analysis).
- **Dimensionality Reduction:** Reducing the number of features while retaining meaningful structure (e.g., PCA, t-SNE).

**Examples of Algorithms:**
- K-Means Clustering
- Hierarchical Clustering
- Principal Component Analysis (PCA)
- DBSCAN

#### Semi-Supervised Learning
Semi-supervised learning lies between supervised and unsupervised learning. It uses a small amount of labeled data along with a larger set of unlabeled data. This is particularly useful when labeling data is expensive or time-consuming.

**Key Characteristics:**
- Combines labeled and unlabeled data.
- Leverages the structure in unlabeled data to improve learning.

**Applications:**
- Text classification with limited labeled examples.
- Image recognition with only a few labeled images.

**Examples of Techniques:**
- Self-training
- Generative Models (e.g., Variational Autoencoders)
- Graph-based semi-supervised learning.
#### Reinforcement Learning
Reinforcement Learning (RL) involves an agent interacting with an environment and learning to make decisions by receiving rewards or penalties based on its actions. The goal is to maximize cumulative rewards.

**Key Characteristics:**
- Focuses on decision-making in dynamic environments.
- No fixed input-output pairs; learning is through trial and error.

**Applications:**
- Game AI (e.g., AlphaGo, Chess engines).
- Robotics (e.g., autonomous navigation).
- Dynamic pricing in e-commerce.

**Examples of Techniques:**
- Q-Learning
- Deep Q-Networks (DQN)
- Policy Gradient Methods
- Actor-Critic Algorithms.


##### Comparison Table

### Data Pre-Processing
Data Pre-Processing refers to cleaning and preparing raw data for analysis by addressing issues like missing values, noise, and inconsistent formats. It ensures the data is in a usable state for feature engineering and model training.

**Key Steps in Data Pre-Processing:**
#### Data Cleaning:
- Handling missing values (e.g., imputation, dropping rows/columns).
- Removing outliers or inconsistent data points.
#### Data Normalization/Scaling:
- Rescaling data to a standard range (e.g., Min-Max scaling, Standardization).
#### Data Integration:
- Combining data from multiple sources into a cohesive dataset.
#### Data Transformation:
- Converting data types, formatting, or applying transformations to improve data consistency.
#### Handling Imbalanced Data:
Using techniques like oversampling, undersampling, or synthetic data generation (e.g., SMOTE) to balance target class distributions.


### Feature Engineering
Feature Engineering is the process of creating new features or modifying existing ones to improve the performance of a machine learning model. It focuses on extracting the most useful information from raw data and making it more predictive or representative of the underlying patterns.

**Key Techniques in Feature Engineering:**
#### Feature Creation:
Generating new features from existing ones (e.g., combining features like speed and time to create distance).
#### Feature Transformation:
Applying mathematical operations (e.g., logarithmic scaling, polynomial features).
#### Feature Encoding:
Converting categorical data into numerical form (e.g., one-hot encoding, label encoding).
#### Feature Selection:
Choosing the most relevant features based on statistical tests, feature importance scores, or correlation analysis.
#### Dimensionality Reduction:
Reducing the number of features while retaining the most relevant information (e.g., PCA, t-SNE).


#### Key Differences
|Aspect|Feature Engineering|Data Pre-Processing|
|:--|:--|:--|
|Objective|To create or enhance features for better predictive power.|To clean, normalize, and prepare raw data for analysis.|
Focus|Improving the quality and relevance of features.|Ensuring the dataset is clean and consistent.|
|Techniques Used|Feature creation, transformation, selection, encoding, dimensionality reduction.|Cleaning, handling missing data, scaling, normalizing, and data type conversions.|
|When Applied|After initial pre-processing and before model training.|Before feature engineering or model training.|
|Tools/Methods|Domain knowledge, statistical methods, algorithms like PCA.|Algorithms for scaling (e.g., StandardScaler), imputation, and cleaning.|
|Outcome|Enhanced feature set for model training.|Usable, clean, and well-structured dataset.|

### Who the outliers are?

```python
outlier_list = []

for count, col in enumerate (data_df.columns):
    
    Q1 = np.percentile(data_df[col], 25)
    Q3 = np.percentile(data_df[col], 75)
    IQR = Q3 - Q1
    ol_dist = IQR * 1.5
    outlier_list.extend(data_df[data_df[col] < Q1 - ol_dist].index)
    outlier_list.extend(data_df[data_df[col] > Q3 + ol_dist].index)

from collections import Counter
outlier_count = Counter(outlier_list)
outlier_count
```

### Remove rows at index 412,414,283

```python
data_df = data_df.drop([412,414,283], axis =0 ).copy()
data_df = data_df[data_df['MEDV'] < 50].copy() # remove where price is 50 and above
data_df.reset_index(inplace=True, drop=True)
data_df.shape
```

### Feature Engineering

```python
colums_sels = ['LSTAT', 'INDUS', 'NOX', 'PTRATIO', 'RM', 'TAX']

scaler = MinMaxScaler()

features_df = data_df[colums_sels].copy()

target = data_df['MEDV']
```

```python
features_df = pd.DataFrame(data=scaler.fit_transform(features_df),
                        columns=colums_sels)

features_df.head()
```

```python
reg = LinearRegression()  # Create an instance of the LinearRegression model.

# Train the linear regression model on the provided data.
# 'features_df' is the input feature matrix (independent variables).
# 'target' is the target vector (dependent variable).
reg.fit(features_df, target)

# Predict the target values using the trained model and the same input features.
# 'y_pred' will store the predicted values for each row in 'features_df'.
y_pred = reg.predict(features_df)
```

```python
plt.scatter(target, y_pred)
plt.plot([y_pred.min(), y_pred.max()], 
         [target.min(), target.max()], 
         c ='r', lw = 3)
```

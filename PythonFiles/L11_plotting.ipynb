```python
import os
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pylab as plt

import seaborn as sns

import plotly.express as px
import plotly.offline as py

import plotly.graph_objects as go


from plotly.figure_factory import create_table
```

```python
## Hyper Parameters
mpl.style.use('ggplot')          #to set style

RANDOM_STATE = 24
np.random.seed(RANDOM_STATE)
```

```python
irisFile = os.path.join('./asp/iris.csv')

iris_features = ['sepal_length',
                 'sepal_width',
                 'petel_length',
                 'petel_width',
                 'target']

iris_df = pd.read_csv(irisFile,skiprows = 1,header= None,
                      names = iris_features)

iris_df.head()
```

```python
#by default plot prints random graph
plt.plot(iris_df.sepal_length,
         iris_df.sepal_width,
         color='r')
plt.show()
```

```python

plt.scatter(iris_df.sepal_length,
         iris_df.sepal_width,
         color='b' ) # k is black
plt.show()
```

```python
plt.scatter(iris_df.sepal_length,
         iris_df.sepal_width,
         color='b', marker='+' ) # k is black
plt.show()
```

```python
colors = ['r','g','b','k']
markers = ['+',',','.','o']
plt.figure(figsize=(15,5))
plt.subplots_adjust(wspace=0.25, hspace=0.25)

for ii in range(3):
    plt.subplot(2,2,ii+1)
    
    col1=iris_df.columns[ii]
    col2=iris_df.columns[ii+1]
    
    plt.scatter(iris_df[col1],iris_df[col2],
                color=colors[ii],marker=markers[ii])
    plt.ylabel(iris_features[ii+1])
    plt.ylabel(iris_features[ii])
    
plt.tight_layout()
```

```python
from matplotlib.patches import Ellipse
x_axis =2
y_axis =3

x_label = iris_features[x_axis]
y_label = iris_features[y_axis]

plt.figure(figsize=(15,12))
ax = plt.subplot(1,1,1, aspect = 'equal')

plt.scatter(iris_df[x_label],iris_df[y_label],
            c=iris_df['target'], cmap='brg', s=50)

plt.xlabel(x_label.capitalize(), fontsize= 10)
plt.ylabel(y_label.capitalize(), fontsize= 10)
plt.title(f'{x_label} vs. {y_label}', fontsize=14,
          fontweight='bold')

plt.annotate('Separate Group',xy=(2,0.25), xytext=(3, 0.25),
             arrowprops = dict(facecolor='gray', shrink= 0.01,
                             connectionstyle='arc3'),fontsize = 12)

ellipse = Ellipse(xy = (4.18,1.6), width = 0.8, height = 0.5, angle=0)
ellipse.set_clip_box(ax.bbox)
ellipse.set_facecolor('m')
ellipse.set_alpha(0.3)
ellipse.set_edgecolor('k')

ax.add_patch(ellipse)

```

```python
binwidth = 0.1

fig, axes = plt.subplots(2,2, figsize=(15,6))

colors = ['r','g','b']
axes = axes.flat

jj = 0

for ax in axes:
    for ii in range(3): #every class 
        start_bin =iris_df[iris_features[jj]].min()
        end_bin = iris_df[iris_features[jj]].max()
        bins = np.arange(start_bin, end_bin, binwidth)
        X1 = iris_df[iris_features[jj]][iris_df['target']== ii]
        ax.hist(X1, bins= bins,
                histtype='bar', 
                color=colors[ii],
                alpha = 0.6)
    ax.set_title(iris_features[jj])
    jj += 1
    
plt.tight_layout()
    
```

```python
n_bin = 10
x = np.random.randn(1000,3)

fig, axes = plt.subplots(2,2)

axes = axes.ravel()

colors = ['r','tan','lime']

ax = axes[0]

ax.hist(x , n_bin , density = True,
        histtype='bar',
        color = colors,
        label = colors)
ax.set_title('Bars with legend')
ax.legend(prop={'size':10})

ax = axes[1]
ax.hist(x , n_bin , density = True,
        histtype='bar',
        color = colors,
        label = colors,stacked = True)
ax.set_title('Stacked bars')

ax = axes[2]
ax.hist(x , n_bin , density = True,
        histtype='step',
        color = colors,
        label = colors,stacked = True,fill = False)
ax.set_title('Stacked bars(no filling)')

ax = axes[3]
x_multi = [np.random.randn(n) for n in [100000,5000,2000]]
ax.hist(x_multi, n_bin , density = True,
        histtype='bar',
        color = colors,
        )
ax.set_title('Diff Sample size')


```

## Box Plot

```python
fig,axes = plt.subplots(2,2,figsize=(15,12)) #for plotting make axes and fix object

axes = axes.ravel()

for count,ax in enumerate(axes):
    
    sns.boxplot(data = iris_df,
                x = iris_features[count],#to horizontal format we x and y for vertical
                hue= 'target', # colour the column target
                ax = ax)
    
plt.tight_layout()
plt.show()
```

```python
fig,axes = plt.subplots(2,2,figsize=(15,12)) #for plotting make axes and fix object

axes = axes.ravel()

for count,ax in enumerate(axes):
    
    sns.violinplot(data = iris_df,
                y = iris_features[count],#to horizontal format we x and y for vertical
                hue= 'target', # colour the column target
                ax = ax)
    
plt.tight_layout()
plt.show()
```
